## grid-template-columns / grid-template-rows

- can use `fr (for fraction), %, rem, px`;
- can use repeat() property ==> repeat(3, 1fr) is the same as 1fr 1fr 1fr
- can also use repeat() property to make it responsive w/o media query eg repeat(auto-fit, minmax(300px, 1fr)) ==> at < 300px, it will fall to 1fr for each column
```
.bento-grid-container {
    display: grid;
    gap: 1rem; /* spacing between cells */

    
    
    grid-template-columns: 100px 100px 200px;
    grid-template-rows: 300px 200px;
}

.bento-grid-item {
    border: 5px solid brown;
    padding: 2rem;
}

## grid-template-areas

- Split areas of rows
- eg code
```
/* to draw out grid area */
.bento-grid-container {
    grid-template-areas:
            'column1 column2 column3'
            'column1 column2 column3'
            'column1 column4 column4'
    ;
}

/* to assign grid area to div*/
.column-one {
    grid-area: column1;
}
```

## place-content: center
- push grid items towards the center of the grid
```
.schedule {
    place-content: center; 
    background-color: var(--clr-Purple-100);
}
```



## CSS Selector

- To target the second child of class inside-element
```
